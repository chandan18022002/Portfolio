```python
import numpy as np 
import cv2 
from numpy import ones,vstack
from numpy.linalg import lstsq
import math
import time 

cap = cv2.VideoCapture(4) 

prev_frame_time = 0
new_frame_time = 0

pt1 = (303, 33)
pt2 = (464, 28)
pt3 = (614, 477)
pt4 = (178, 479)
pts=[pt1,pt2,pt3,pt4]
pts_x=[]
pts_y=[]
for i in pts:
    pts_x.append(i[1])
    pts_y.append(i[0])
pts_x.sort()
pts_y.sort()

def mask_crop(img,pt1,pt2,pt3,pt4):
    y,x,_ = img.shape
    mask = np.zeros((y,x), np.uint8)
    triangle_cnt = np.array( [pt4, pt1, pt2, pt3] )
    cv2.drawContours(mask, [triangle_cnt], 0, 255, -1)
    img = cv2.bitwise_and(img, img, mask=mask)
    return img

def mask_blu(img):
    image=cv2.blur(img, (6,6))
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, (90, 100, 100), (135, 250, 150))
    target = cv2.bitwise_and(img,img, mask=mask)
    return target

def mask_pnk(img):
    image=cv2.blur(img, (6,6))
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, (150, 100, 100), (170, 200, 200))
    target = cv2.bitwise_and(img,img, mask=mask)
    return target

def mask_yel(img):
    image=cv2.blur(img, (4,4))
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, (27, 38, 100), (36, 200, 200))
    target = cv2.bitwise_and(img,img, mask=mask)
    return target

def mask_org(image):
    image=cv2.blur(image, (4,4))
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, (0, 70, 50), (15, 255, 255))
    target = cv2.bitwise_and(img,img, mask=mask)
    return target

while(cap.isOpened()): 
    ret, frame = cap.read()
    new_frame_time = time.time() 
    if ret:
        img=frame
        img=mask_crop(img,pt1,pt2,pt3,pt4)
        img=img[pts_x[0]:pts_x[-1],pts_y[0]:pts_y[-1],:]
        
        gray = mask_org(img)
	     
        font = cv2.FONT_HERSHEY_SIMPLEX 
        fps = 1/(new_frame_time-prev_frame_time) 
        prev_frame_time = new_frame_time 
        fps = int(fps) 
        fps = str(fps) 
        cv2.putText(gray, fps, (7, 70), font, 3, (100, 255, 0), 3, cv2.LINE_AA) 
        cv2.imshow('o_frame',img) 
        cv2.imshow('frame', gray) 
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

cap.release() 
cv2.destroyAllWindows() 
```
